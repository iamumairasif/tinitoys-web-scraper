```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://tinitoys.bg/index.php?route=product/category&path=129"
```

```python
response = requests.get("https://tinitoys.bg/index.php?route=product/category&path=129")
print(response)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
links = []
for a_tag in soup.find_all("a", href=True):
    links.append(a_tag['href'])
```

```python
for link in links:
    print(link)
```

```python
data = []

page = 1

while True:

    print("Opening Page", page)

    url = url + "&page=" + str(page)
    response = requests.get(url)

    soup = BeautifulSoup(response.text, "html.parser")

    products = soup.find_all("div", class_="product-layout")

    if len(products) == 0:
        break

    for product in products:

        name = product.find("div", class_="name").text

        code = product.find("span", class_="stat-2").find_all("span")[1].text

        price = product.find("span", class_="price-normal").text

        link = product.find("a", class_="product-img")["href"]

        data.append([code, name, price, link])

    page = page + 1
```

```python
df = pd.DataFrame(data, columns=["Product Code", "Name", "Price", "Link"])

df.to_csv("products.csv", index=False)

print(df.head())

print("Total Products:", len(df))
```

```python
import json

with open("products.json", "w") as file:
    json.dump(data, file, indent=4)

print("JSON file created")
```
